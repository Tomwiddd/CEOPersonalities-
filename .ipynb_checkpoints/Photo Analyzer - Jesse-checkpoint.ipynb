## Photo Analyzer Test

```python
import pandas as pd
import numpy as np
from deepface import DeepFace
```

```python
ceo_data = pd.read_csv('ceo_data.csv')

ceo_data
```

```python

# Analyze the image
result = DeepFace.analyze(img_path="headshot.jpg", actions=['age', 'gender', 'race', 'emotion'], detector_backend='opencv')

# If result is a list, get the first item
data = result[0] if isinstance(result, list) else result

# Flatten the nested dictionaries
flat_data = {
    'age': data['age'],
    'dominant_gender': data['dominant_gender'],
    'dominant_race': data['dominant_race'],
    'dominant_emotion': data['dominant_emotion'],
    **data['gender'],
    **data['race'],
    **data['emotion']
}

# Create DataFrame
df = pd.DataFrame([flat_data])


df
```
