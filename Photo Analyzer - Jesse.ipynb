```python
# Folder containing subfolders with images
base_dir = 

# List to store all analysis results
all_data = []

# Loop through all folders and files
for root, dirs, files in os.walk(base_dir):
    for file in files:
        if file.lower().endswith(('.jpg', '.jpeg', '.png')):
            img_path = os.path.join(root, file)
            try:
                result = DeepFace.analyze(img_path=img_path,
                                          actions=['age', 'gender', 'race', 'emotion'],
                                          detector_backend='opencv')
                data = result[0] if isinstance(result, list) else result

                # Flatten nested results
                flat_data = {
                    'file_path': img_path,
                    'age': data['age'],
                    'dominant_gender': data['dominant_gender'],
                    'dominant_race': data['dominant_race'],
                    'dominant_emotion': data['dominant_emotion'],
                    **data['gender'],
                    **data['race'],
                    **data['emotion']
                }

                all_data.append(flat_data)
            except Exception as e:
                print(f"Error processing {img_path}: {e}")

# Create a DataFrame of all results
df = pd.DataFrame(all_data)

# Save to CSV (optional)
df.to_csv("face_analysis_results.csv", index=False)

# Display the DataFrame
print(df)
```
