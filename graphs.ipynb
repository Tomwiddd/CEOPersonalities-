```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("outputs/output_daily.csv")

df_clean = df.dropna(subset=['Return', 'CEO', 'dominant_gender', 'dominant_race', 'dominant_emotion'])

gender_returns = df_clean.groupby('dominant_gender')['Return'].mean()

race_returns = df_clean.groupby('dominant_race')['Return'].mean()

emotion_returns = df_clean.groupby('dominant_emotion')['Return'].mean()

tenure_returns_gender = df_clean.groupby('dominant_gender')['Tenure_Cum_Ret_Overall'].mean()

year_returns_emotion = df_clean.groupby('dominant_emotion')['Year_Cum_Ret_Overall'].mean()

avg_age_gender = df_clean.groupby('dominant_gender')['Age'].mean()

gender_counts = df_clean['dominant_gender'].value_counts()


def plot_bar(data, title, xlabel, ylabel, color='skyblue'):
    data.plot(kind='bar', color=color)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.tight_layout()
    plt.show()

plot_bar(gender_returns, 'Average Daily Return by Gender', 'Gender', 'Return')
plot_bar(race_returns, 'Average Daily Return by Race', 'Race', 'Return', 'salmon')
plot_bar(emotion_returns, 'Average Daily Return by Emotion', 'Emotion', 'Return', 'seagreen')
plot_bar(tenure_returns_gender, 'Tenure Cumulative Return by Gender', 'Gender', 'Cumulative Return', 'steelblue')
plot_bar(year_returns_emotion, 'Yearly Cumulative Return by Emotion', 'Emotion', 'Cumulative Return', 'coral')
plot_bar(avg_age_gender, 'Average CEO Age by Gender', 'Gender', 'Age', 'gold')
```
